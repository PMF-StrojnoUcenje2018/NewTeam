```python
import numpy as np
import pandas as pd
import pickle
import os
import read_jsons
```

Uƒçitavanje testnih podataka
---

```python
exe_names = []

with open('../data/test/sample_sha1_list.txt') as f:
    for line in f:
        exe_names.append(line.rstrip())
        
N = len(exe_names)
print("Broj datoteka: " + str(N))

file_list = [os.path.join('../data/test/reports', li[-2:], li+'.json') for li in exe_names]
```

```python
df = read_jsons.get_dataframe(file_list)
df
```

```python
X = df

names = X['name']
X.drop('name',  axis=1, inplace=True)
#naknadno izbaƒçeni featuresi
X.drop('log_max_heap_reserve',     axis=1, inplace=True)
X.drop('log_repeated_entry_sizes', axis=1, inplace=True)
X.drop('non_empty_metadatas',      axis=1, inplace=True)
X.drop('certificate',              axis=1, inplace=True)
X.drop('cert_malformed',           axis=1, inplace=True)
X.drop('rich_header_valid',        axis=1, inplace=True)
X.drop('max_opt_header_size',      axis=1, inplace=True)
X
```

Uƒçitavanje modela
---

```python
#Load model
filename = 'TrainedModel.sav'
model = pickle.load(open(filename, 'rb'))
```

```python
#Predict
predicted = model.predict(X)

result = {
    "packer type": names,
    "is packer": predicted.astype(int),
}

resultf = pd.DataFrame(result, columns = result)
resultf.to_csv('result.tsv', header = False, index = False,sep='\t')
#broj_testnih_primjeraka = len(Y)
#print('Broj test primjeraka: ' + str(len(Y)))
#broj_tocno_klasificiranih = sum(predicted == Y)
#print('Broj toƒçno klasificiranih: ' + str(broj_tocno_klasificiranih))
#tocnost = broj_tocno_klasificiranih/broj_testnih_primjeraka
#print('Toƒçnost: '+str(tocnost*100)+'%')
```
