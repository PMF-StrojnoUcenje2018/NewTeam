```python
import os
import pandas as pd
import numpy as np
import read_jsons
```

Učitavanje imena datoteka i tipova packera
---

```python
exe_names = []
exe_packer_types = []

with open('../data/training/samples_by_packer_group.txt') as f:
    for line in f: #.readlines(::20):
        spl = line.split(',')
        name = spl[0]
        packer_type = spl[1]
        exe_names.append(name)
        exe_packer_types.append('t'+str(int(packer_type)))

exe_is_packer = [packer_type != 't0' for packer_type in exe_packer_types]
```

```python
N = len(exe_names)
print("Broj datoteka: " + str(N))
```

```python
print("Broj različitih tipova pakera: \n" +
      str({i: sum(1 if exe_packer_types[j] == 't'+str(i) else 0 for j in range(len(exe_packer_types))) for i in range(5)}))
```

Stvaranje DataFrame-a
---

```python
file_list = [os.path.join('../data/training/reports', li[-2:], li+'.json') for li in exe_names]
```

```python
%%time
df = read_jsons.get_dataframe(file_list)
```

```python
data = {
    "packer type": exe_packer_types,
    "is packer": exe_is_packer,
}

df2 = pd.DataFrame(data, columns = data)

df = df.join(df2)
df
```
